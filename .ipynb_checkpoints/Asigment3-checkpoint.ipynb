```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib notebook
```

```python
def crear_tablero():
    fig, ax = plt.subplots()
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.set_aspect('equal')
    ax.set_xticks(np.arange(0.5, 8.5, 1))
    ax.set_yticks(np.arange(7.5, -0.5, -1))  # NumeraciÃ³n de arriba hacia abajo
    ax.set_xticklabels(list("ABCDEFGH"))
    ax.set_yticklabels(list("12345678"))

    for i in range(8):
        for j in range(8):
            rect = Rectangle((i, j), 1, 1, facecolor='green', edgecolor='black')
            ax.add_patch(rect)

    return fig, ax

current_color = 'white'  # Inicialmente, el color es azul

def pintar_casilla(event, fig, ax):
    global current_color
    if event.xdata is not None and event.ydata is not None:
        x, y = int(event.xdata), int(event.ydata)
        color = 'white' if current_color == 'black' else 'black'
        rect = Rectangle((x, y), 1, 1, facecolor=color, edgecolor='black')
        ax.add_patch(rect)
        fig.canvas.draw()
        current_color = color  # Cambiar el color

if __name__ == "__main__":
    fig, ax = crear_tablero()
    fig.canvas.mpl_connect('button_press_event', lambda event: pintar_casilla(event, fig, ax))
    plt.show()
```
